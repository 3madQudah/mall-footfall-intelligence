```python
# This Charts show the null values in the dataset.
# barplot is used to show the count of null values in each column.
# And show the percentage of null values in each column.
plt.figure(figsize=(10, 6))
sns.barplot(x=Data.isnull().sum().index, y=Data.isnull().sum().values)
plt.title("Count of Null Values in Each Column")
plt.xlabel("Columns")
plt.ylabel("Count of Null Values")
plt.xticks(rotation=45)
plt.show()
```
